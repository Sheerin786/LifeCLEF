```python

import os
import shutil

# Define the main folder path
main_folder = "/home/sheerin/AnimalCLEF/Train"

# Walk through all subdirectories
for root, dirs, files in os.walk(main_folder):
    if root != main_folder:  # Avoid moving files already in the main folder
        for file in files:
            file_path = os.path.join(root, file)  # Get full path of the file
            new_path = os.path.join(main_folder, file)  # Destination path

            # Move file to the main folder
            shutil.move(file_path, new_path)
            print(f"Moved: {file_path} -> {new_path}")

print("All files have been moved to the main folder successfully.")
```

```python
import csv
from pathlib import Path
csv_file = "/home/sheerin/AnimalCLEF/metadata.csv"
updated_rows = []

# Read and modify the CSV file
with open(csv_file, "r") as file:
    reader = csv.reader(file)
    header = next(reader)  # Store the header

    for row in reader:
        if len(row) >= 3:  #
            print(row[2])
            path = Path(row[2])
            filename = path.name
            row[2] = filename  # Replace with a generated value
        updated_rows.append(row)

# Write back the modified data
with open(csv_file, "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(header)  # Write the header back
    writer.writerows(updated_rows)  # Write the updated rows

print("Third column values have been updated successfully.")

#AnimalCLEF/Trial.csv
```

```python
import os
import csv

# Define folder path where images are stored
folder_path = "/home/sheerin/AnimalCLEF/Train"
csv_file = "/home/sheerin/AnimalCLEF/metadata.csv"
i=1
# Read the CSV file
with open(csv_file, "r") as file:
    reader = csv.reader(file)
    next(reader)  # Skip the header if there is one

    for row in reader:
        if len(row) >= 3:  # Ensure enough columns exist
            old_name = row[2]  # File name from third column
            #print(old_name)
            new_name = f"{row[1]}_{i}+.jpg"  # New name from second column
            #print(new_name)
            i=i+1
            old_path = os.path.join(folder_path, old_name)
            new_path = os.path.join(folder_path, new_name)
            #print(old_path)

            # Check if the file exists before renaming
            if os.path.exists(old_path):
                os.rename(old_path, new_path)
                print(f"Renamed: {old_name} -> {new_name}")
            else:
                print(f"File not found: {old_name}")


import os

folder = "images"
filename = "image.jpg"
i = 1

# Generate a new filename if it already exists
while os.path.exists(os.path.join(folder, filename)):
    filename = f"image_{i}.jpg"
    i += 1

print("Saving as:", filename)
```
